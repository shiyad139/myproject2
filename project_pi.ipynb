```python
# Import library
from sklearn.linear_model import LogisticRegression

# Sample training data
# Format: [Glucose, BMI]
X = [
    [85, 25],
    [90, 28],
    [150, 35],
    [160, 40]
]

# Output (0 = Not diabetic, 1 = Diabetic)
y = [0, 0, 1, 1]

# Create model
model = LogisticRegression()

# Train model
model.fit(X, y)

# Take user input
glucose = float(input("Enter Glucose level: "))
bmi = float(input("Enter BMI: "))

# Predict
prediction = model.predict([[glucose, bmi]])

# Output result
if prediction[0] == 0:
    print("Not Diabetic")
else:
    print("Diabetic")
```
